## 1. Finding Key Connectors

```python
# Identify who the “key connectors” are among data scientists
users = [
{ "id": 0, "name": "Hero" },
{ "id": 1, "name": "Dunn" },
{ "id": 2, "name": "Sue" },
{ "id": 3, "name": "Chi" },
{ "id": 4, "name": "Thor" },
{ "id": 5, "name": "Clive" },
{ "id": 6, "name": "Hicks" },
{ "id": 7, "name": "Devin" },
{ "id": 8, "name": "Kate" },
{ "id": 9, "name": "Klein" }
]
```

```python
users
```

```python
friendships = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4), (4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)]
```

```python
## the data scientist with id 0 (Hero) and the data scientist with id 1 (Dunn) are friends
friendships
```

```python
from IPython.display import Image
Image("./DSnetwork.png")
```

```python
# We might want to add a list of friends to each user. First we set each user’s friends property to an empty list:
for user in users:
    user["friends"] = []
```

```python
users
```

```python
for i, j in friendships:
    # this works because users[i] is the user whose id is i
    users[i]["friends"].append(users[j]) # add i as a friend of j
    users[j]["friends"].append(users[i]) # add j as a friend of i
```

```python
users[0]
```

#### What’s the average number of connections?

```python
def number_of_friends(user):
    """how many friends does _user_ have?"""
    return len(user["friends"]) # length of friend_ids list
```

```python
total_connections = sum(number_of_friends(user) for user in users)
```

```python
total_connections
```

```python
from __future__ import division # 실수형 변환
num_users = len(users)
avg_connections = total_connections / num_users
```

```python
avg_connections
```

#### Sort them from “most friends” to “least friends”

```python
num_friends_by_id =  [(user['id'], number_of_friends(user)) for user in users]
num_friends_by_id
```

```python
# id 별 친구 수 파악하기

num_friends_by_id = []
for user in users:
    num_friends_by_id.append((user['id'], number_of_friends(user)))
```

## 2. Data Scientists You May Know

```python
def friends_of_friend_ids_bad(user):
    # "foaf" is short for "friend of a friend"
    return [foaf['id'] 
            for friend in user['friends']  # for each of user's friends
            for foaf in friend["friends"]]   # get each of _their_ friends
```

```python
friends_of_friend_ids_bad(users[0])
```

```python
from collections import Counter # not loaded by default

def not_the_same(user, other_user):
    """two users are not the same if they have different ids"""
    return user["id"] != other_user["id"]

def not_friends(user, other_user):
    """other_user is not a friend if he's not in user["friends"];
    that is, if he's not_the_same as all the people in user["friends"]"""
    return all(not_the_same(friend, other_user) for friend in user["friends"])

def friends_of_friend_ids(user):
    return Counter(foaf["id"] 
                            for friend in user["friends"] # for each of my friends
                            for foaf in friend["friends"] # count *their* friends
                            if not_the_same(user, foaf) # who aren't me
                            and not_friends(user, foaf)) # and aren't my friends

print friends_of_friend_ids(users[3]) # Counter({0: 2, 5: 1}) 
'''
This correctly tells Chi (id 3) that she as two mutual friends with Hero (id 0) but only one mutual friend with Clive (id 5).
'''
```

## 3. Salaries and Experience

```python
# Each user’s salary (in dollars) and tenure as a data scientist (in years)

salaries_and_tenures = [(83000, 8.7), (88000, 8.1),
(48000, 0.7), (76000, 6),
(69000, 6.5), (76000, 7.5),
(60000, 2.5), (83000, 10),
(48000, 1.9), (63000, 4.2)]
```

```python
salaries_and_tenures
```

```python
# keys are years, values are lists of the salaries for each tenure
salary_by_tenure = defaultdict(list)
```

```python
for salary, tenure in salaries_and_tenures:
    salary_by_tenure[tenure].append(salary)

salary_by_tenure
```

```python
# keys are years, each value is average salary for that tenure
average_salary_by_tenure = { tenure : sum(salaries) / len(salaries)
                                            for tenure, salaries in salary_by_tenure.items()
}
```

```python
average_salary_by_tenure
```

```python
def tenure_bucket(tenure):
    if tenure < 2:
        return "less than two"
    elif tenure < 5:
        return "between two and five"
    else:
        return "more than five"
```

```python
# keys are tenure buckets, values are lists of salaries for that bucket
salary_by_tenure_bucket = defaultdict(list)
```

```python
for salary, tenure in salaries_and_tenures:
    bucket = tenure_bucket(tenure)
    salary_by_tenure_bucket[bucket].append(salary)
```

```python
salary_by_tenure_bucket
```

```python
# keys are tenure buckets, values are average salary for that bucket
average_salary_by_bucket = {tenure_bucket : sum(salaries) / len(salaries)
                                            for tenure_bucket, salaries in salary_by_tenure_bucket.iteritems()
}
```

```python
average_salary_by_bucket
```

```python
def predict_paid_or_unpaid(years_experience):
    if years_experience < 3.0:
        return "paid"
    elif years_experience < 8.5:
        return "unpaid"
    else:
        return "paid"
```

```python
for salary, tenure in salaries_and_tenures:
    predict = predict_paid_or_unpaid(tenure)
    print str(tenure) + ' ' + str(predict)
```
